```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv('../Files/linear_Regression/canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.scatter(df['year'],df['per capita income (US$)'],color='green',marker='^')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Canada\'s per capita income')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```
